```python
import numpy as np
from cv2 import imread
import matplotlib.pyplot as plt
```

```python
img= imread('mandril.png')
#img = imread('scale.png')
#img = cv2.imread ('mandril.png') # GBR
```

```python
plt.imshow(img)
```

```python
def howis(img):
    print('size: ', img.shape)
    print('max: ', np.max(img))
    print('min: ', np.min(img))
```

```python
X=img[:,:,0]
howis(X)
```

```python
#mmuestreo

#Construccin de versin ms pequea de la imagen original (escala de grises), toma un pixel dada determinado paso
# o distancia y ese pixeel se asigna a otro espacio (matriz)

d=8
(Nx,Mx) = X.shape
#Nx num. de filas
#Ny numero de columnas

#range (start, stop, d)

ix = range(0,Nx,d) #indices de las posiciones que queremos muestrear
jx=range(0,Mx,d) # 0 d 2d 3d 4d

Ny=len(jx) #new tamao de la imagen muestreada
My=len(ix) #new tamao de la imagen muestreada

Y= np.zeros((Ny,My),np.uint8) #Crea imagen de salida (con tamao reducido Ny y My)

for i in range (Ny): #recorre filas
    for j in range(My): #recorre columnas
        Y[i,j]=X[ix[i],jx[j]] # para cada pixel de Y[i,j], toma el pixel de la imagen original X en la posicin (fila, columna) =  ()

plt.imshow(Y, cmap='gray')
plt.show()
howis(Y)
```

##Submuestreo pero conservando TAMAO ORIGINAL (512,512) quitarle un poco de calidad pero se conserva tamao como en Instagram

```python
d=8 #tamao del paso

#Tupla que nos devuelva tamao del original
(Nx,Mx) = X.shape

Y= np.zeros((Nx,Mx), dtype=np.uint8)

for i in range(Nx):
    for j in range(Mx):
        ix = (i//d) * d # para redondear sin decimales se usa //, devuelve ENTEROS #VALOR ANCLA
        jx = (j//d) * d #VALOR ANCLA
        Y[i,j]=X[ix,jx] #asigna el valor en el posicin i,j de valoor ancla

#para i... 0 a 7... d=0
#para  i de 9 al 15... d=1...

plt.imshow(Y, cmap='gray')
plt.show()
howis(Y)
```

## Histograma:

```python
def imhist(X,n):
    (N,M)=X.shape
    h =np.zeros((n,))
    for i in range(N):
        for j in range(M):
            x = X[i,j]
            h[x] = h[x]+1
    return h
```

```python
n=256
h= imhist(X,n)
plt.plot(figsize=(10,5))
plt.plot(range(n),h[0:n])
plt.show()
```
